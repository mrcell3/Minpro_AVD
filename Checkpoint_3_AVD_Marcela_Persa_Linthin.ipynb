### Outliers

Outliers adalah nilai yang jauh berbeda dari nilai lainnya dalam dataset. Nilai Outlier bisa jauh lebih rendah atau lebih tinggi. Outlier bisa terjadi karena berbagai alasan seperti faktor kesalahan maupun kejadian lain yang tidak terduga.

Melakukan pengecekan outliers

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

# **VISUALISASI**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. Distribusi metode pembayaran
plt.figure(figsize=(8, 5))
sns.countplot(x="Payment", data=df, palette="pastel")
plt.title("Distribusi Metode Pembayaran")
plt.xlabel("Metode Pembayaran")
plt.ylabel("Jumlah Transaksi")
plt.show()
```

Insight: Mengetahui metode pembayaran yang paling sering digunakan pelanggan.

- Metode pembayaran E-wallet dan Cash memiliki jumlah transaksi yang hampir sama dan menjadi yang paling banyak digunakan.
- Credit Card sedikit lebih rendah penggunaannya dibandingkan metode lainnya.

Tindakan: Jika ingin meningkatkan penggunaan Credit Card, supermarket bisa menawarkan promo atau diskon khusus untuk transaksi menggunakan kartu kredit.
E-wallet dan Cash bisa tetap difasilitasi dengan baik karena banyak digunakan oleh pelanggan.

```python
# 2. Pola penggunaan metode pembayaran berdasarkan waktu transaksi
plt.figure(figsize=(10, 5))
sns.histplot(data=df, x="Hour", hue="Payment", multiple="stack", bins=24, palette="pastel")
plt.title("Pola penggunaan metode pembayaran berdasarkan waktu transaksi")
plt.xlabel("Jam Transaksi")
plt.ylabel("Jumlah Transaksi")
plt.xticks(range(0, 24))
plt.grid()
plt.show()
```

Insight: Mengetahui pola metode pembayaran di waktu tertentu.

- Puncak transaksi terjadi pada siang hingga malam hari (10:00 - 20:00).
- E-wallet lebih dominan dibandingkan metode lain, terutama pada jam-jam sibuk.
- Cash dan Credit Card memiliki pola penggunaan yang cukup stabil sepanjang hari.

Tindakan: Jika ingin meningkatkan efisiensi checkout di jam sibuk, supermarket bisa meningkatkan jumlah kasir atau menyediakan self-checkout untuk E-wallet.
Promo khusus pada jam tertentu bisa diberikan untuk metode pembayaran tertentu (misalnya, diskon tambahan jika menggunakan kartu kredit setelah pukul 18:00).
